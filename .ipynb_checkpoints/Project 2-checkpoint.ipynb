# Problem Statement:

**Predictive Modeling & Plan**
- We built a series of predictive models starting with a simple SLR which provides us a benchmark from which we hope to show the impact of MLR and machine learning on improved predictive performance with each iteration.  
- Target: 'SalePrice' from one or more selected 'features'.
- We iterated through 4 modeling methods & left one additional Method for Future analysis:

    1. Method 1. Run a simple SLR on the best selected Continuous feature, which can provide a benchmark for the other models to beat.
    2. Method 2. Run a MLR using (1) plus other selected Continuous features provided.
    3. Method 3. Run a MLR using (2) plus selected binned Ordinal features
    4. Method 4. Run a MLR using (3) plus selected dummified Nominal features.
    5. Future Method 5. Run a Lasso on all other remaining features to see what the model chooses as potentially important, then add these features to Method (4) in developing a final model.  
- A successful model may allow us to confidently predict Property prices in the test and real world datasets.
- A successful model can be check for performance accuracy using Cross Validation scores.   


# Conclusions and Recommendations:

## ---Model------R^2-------RSME---Conclusion:
### -----SLR---------0.547--------50,382---Easily computed & easily explained. Lowest test scores.
### -----MLR10------0.835--------32,261---Materially better test scores, but at the expense of explainability. 
### -----MLR12------0.858--------29,571---Modest improvement in test scores, and incresingly complex.
### -----MLR30------0.868--------28,502---Minor improvement in test scoring, much more complex.

- With over 80 possible features our model had the potential to be too wide. As such, our approach was to start with the best correlation as a SLR and then build into more complex MLR for less incremental returns to model prediction.

- We actually like and appreciate our SLR model for its ease of use and explainability to be used by professionals 'in-the-field'. Its RMSE of 57,182 is probably acceptable as a 'first-cut'.

- However, for more rigorous analysis we can have greater confidence in our Model 3 which uses 12 features to predict SalePrice. Its RMSE is much better at just 29,571, but what it offers in-terms of predictive power we lose in-terms of ease-of-use and explainability.

```python
# this is a function used to calculate linear regression that we found on the internet.

def lin_reg(x,y):
    beta_1 = (scipy.stats.pearsonr(x,y)[0])*(np.std(y)/np.std(x))
    
    # Pearson Co. Coef returns a tuple so it needs to be sliced/indexed
    # the optimal beta is found by: mean(y) - b1 * mean(x)
    beta_0 = np.mean(y)-(beta_1*np.mean(x)) 
    
    #Print the Optimal Values
    print ('The Optimal Y Intercept is ', beta_0)
    print ('The Optimal slope is ', beta_1)
```

```python
# this passes our data to the SLR function for calculation and interpretation.
# From that we can calculate a the simple predictive SalePrice table below.

x = df_train_continuous['Gr Liv Area'].values
y = df_train_continuous['SalePrice'].values
lin_reg(x,y)

# The y intercept predicts that our 'SalePrice' starts at around $5,841 for a Gr Liv Area of around 0.
# Whilst the slope predicts that for every 1-unit increase in 'Gr Liv Area' 
#   the 'SalePrice' increases by $117

# Note: Gr Liv Area (Continuous): Above grade (ground) living area square feet
```

```python
# this is a function used to calculate our Standard Scaler that we found on the internet.

def standardize(X_train, X_test):
    ss = StandardScaler()
    X_train_sc = pd.DataFrame(ss.fit_transform(X_train))
    X_test_sc = pd.DataFrame(ss.transform(X_test))
    
    return X_train_sc, X_test_sc
```

```python
# this is a function used to calculate our Results Summary Scores that we found on the internet.

def test_data(model, X, y):
    x1, x2, y1, y2 = train_test_split(X, y, random_state=42)
    model.fit(x1, y1)
    train_pred = model.predict(x1)
    test_pred = model.predict(x2)
    print('Training Results')
    print('R2: ' + str(round(model.score(x1, y1), 3)))
    print('RMSE: ' + str(round(np.sqrt(mean_squared_error(y1, train_pred)), 3)))
    print()
    print('Testing Results')
    print('R2: ' + str(round(model.score(x2, y2), 3)))
    print('RMSE: ' + str(round(np.sqrt(mean_squared_error(y2, test_pred)), 3)))
```
